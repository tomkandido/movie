```python
# Função que adiciona um novo contato à agenda
def adicionar_contato(agenda):
    nome = input("Digite o nome do contato: ")
    telefone = input("Digite o telefone do contato: ")
    email = input("Digite o e-mail do contato: ")
    twitter = input("Digite o usuário do Twitter do contato (ou deixe em branco): ")
    instagram = input("Digite o usuário do Instagram do contato (ou deixe em branco): ")
    agenda[nome] = {'telefone': telefone, 'email': email, 'twitter': twitter, 'instagram': instagram}
    print("Contato adicionado com sucesso!")

# Função que consulta um contato na agenda
def consultar_contato(agenda):
    nome = input("Digite o nome do contato que deseja consultar: ")
    if nome in agenda:
        print("Nome:", nome)
        print("Telefone:", agenda[nome]['telefone'])
        print("E-mail:", agenda[nome]['email'])
        print("Twitter:", agenda[nome]['twitter'])
        print("Instagram:", agenda[nome]['instagram'])
    else:
        print("Contato não encontrado na agenda.")

# Função que remove um contato da agenda
def remover_contato(agenda):
    nome = input("Digite o nome do contato que deseja remover: ")
    if nome in agenda:
        del agenda[nome]
        print("Contato removido com sucesso!")
    else:
        print("Contato não encontrado na agenda.")

# Função que altera os dados de um contato na agenda
def alterar_contato(agenda):
    nome = input("Digite o nome do contato que deseja alterar: ")
    if nome in agenda:
        print("Digite os novos dados:")
        telefone = input("Digite o novo telefone do contato (ou deixe em branco para manter o mesmo): ")
        if telefone:
            agenda[nome]['telefone'] = telefone
        email = input("Digite o novo e-mail do contato (ou deixe em branco para manter o mesmo): ")
        if email:
            agenda[nome]['email'] = email
        twitter = input("Digite o novo usuário do Twitter do contato (ou deixe em branco para manter o mesmo): ")
        if twitter:
            agenda[nome]['twitter'] = twitter
        instagram = input("Digite o novo usuário do Instagram do contato (ou deixe em branco para manter o mesmo): ")
        if instagram:
            agenda[nome]['instagram'] = instagram
        print("Contato alterado com sucesso!")
    else:
        print("Contato não encontrado na agenda.")

# Função que gera um relatório com os dados de todos os contatos
def gerar_relatorio(agenda):
    print("Relatório de Contatos:")
    print("{:<20} {:<15} {:<30} {:<20} {:<20}".format("Nome", "Telefone", "E-mail", "Twitter", "Instagram"))
    for nome, dados in agenda.items():
        print("{:<20} {:<15} {:<30} {:<20} {:<20}".format(nome, dados['telefone'], dados['email'], dados['twitter'], dados['instagram']))

# Função que salva os dados da agenda em um arquivo
def salvar_agenda(agenda, arquivo):
    with open(arquivo, 'w') as f:
        for nome, dados in agenda.items():
            linha = ",".join([nome, dados['telefone'], dados['email'], dados['twitter'], dados['instagram']]) + "\n"
            f.write(linha)

# Função que carrega os dados da agenda a partir de um arquivo
def carregar_agenda(arquivo):
    agenda = {}
    with open(arquivo, 'r') as f:
        for linha in f:
            nome, telefone, email, twitter, instagram = linha.strip().split(",")
            agenda[nome] = {'telefone': telefone, 'email': email, 'twitter': twitter, 'instagram': instagram}
    return agenda

# Função para o menu principal
def menu():
    print("Bem-vindo à Agenda Telefônica!")
    agenda = {}
    while True:
        print("\nOpções:")
        print("1. Adicionar Contato")
        print("2. Consultar Contato")
        print("3. Remover Contato")
        print("4. Alterar Contato")
        print("5. Gerar Relatório")
        print("6. Salvar Agenda")
        print("7. Carregar Agenda")
        print("8. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == '1':
            adicionar_contato(agenda)
        elif opcao == '2':
            consultar_contato(agenda)
        elif opcao == '3':
            remover_contato(agenda)
        elif opcao == '4':
            alterar_contato(agenda)
        elif opcao == '5':
            gerar_relatorio(agenda)
        elif opcao == '6':
            arquivo = input("Digite o nome do arquivo para salvar a agenda: ")
            salvar_agenda(agenda, arquivo)
        elif opcao == '7':
            arquivo = input("Digite o nome do arquivo para carregar a agenda: ")
            agenda = carregar_agenda(arquivo)
        elif opcao == '8':
            print("Obrigado por utilizar a Agenda Telefônica!")
            break
        else:
            print("Opção inválida. Tente novamente.")

# Executando o menu principal
menu()


# Criando alguns contatos de exemplo
agenda = {
    'João': {'telefone': '123456789', 'email': 'joao@email.com', 'twitter': '@joao', 'instagram': 'joao_insta'},
    'Maria': {'telefone': '987654321', 'email': 'maria@email.com', 'twitter': '@maria', 'instagram': 'maria_insta'}
}

# Adicionando mais contatos
adicionar_contato(agenda)
adicionar_contato(agenda)

# Consultando um contato
consultar_contato(agenda)

# Removendo um contato
remover_contato(agenda)

# Alterando um contato
alterar_contato(agenda)


```
